# Bayes Factor Calculations with Dynesty
In this notebook, I demonstrate how we do our calculations of the Bayesian evidence (and by extension the Bayes factor) for a spectrum. 

```python
from astropy.io import fits
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.pyplot as plt
import dynesty
from dynesty import plotting as dyplot
import ipyparallel as ipp
```

```python
# Read in Fits
spectrum = fits.open('ReferenceSpectra/Reference-Spectrum-R5000.fits')
header = spectrum[0].header
spec = spectrum[1].data
channel = []
counts = []
for chan in spec:  # Only want SN3 region
    channel.append(chan[0])
    counts.append(chan[1])
counts = np.array(counts)#[600:750]
y = counts#[600:750]
x = np.linspace(-np.argmax(counts),len(counts)-np.argmax(counts), len(counts))
plt.plot(x, y)
plt.show()
```

## Define Model, Likelihood, and Priors
We define our two models: a single Gaussian and a double Gaussian. We also define our likelihood function and assume a constant error. We assume a Gaussian likelihood function. Finally, we assume non-informative priors.

```python
def gauss(channel, A, sigma, x0):
    return A*np.exp(-(x - x0)**2/(2*sigma**2))

def gauss2(channel, A, sigma, x0, A2, sigma2_, x20):
    return A*np.exp(-(x - x0)**2/(2*sigma**2)) + A2*np.exp(-(x - x20)**2/(2*sigma2_**2))

def loglike1(theta):
    A_, sigma_, x0_ = theta
    model_eval = gauss(x, A_, sigma_, x0_)
    sigma2 = 1e-4#yerr ** 2 
    return -0.5 * np.sum((y - model_eval) ** 2 / sigma2 + np.log(2*np.pi*sigma2))

def loglike2(theta):
    A_, sigma_, x0_, A2_, sigma2_, x20_ = theta
    model_eval = gauss2(x, A_, sigma_, x0_, A2_, sigma2_, x20_)
    sigma2 = 1e-4#yerr ** 2 
    return -0.5 * np.sum((y - model_eval) ** 2 / sigma2 + np.log(2*np.pi*sigma2))

# prior transform
def prior_transform1(utheta):
    uA, usigma, ux0 = utheta
    A = 3*uA
    sigma = 10*usigma
    x0 = 20*(ux0-0.5)
    
    return A, sigma, x0

# prior transform
def prior_transform2(utheta):
    uA, usigma, ux0, uA2, usigma2, ux20 = utheta
    A = 3*uA
    sigma = 10*usigma
    x0 = 20*(ux0-0.5)
    A2 = 3*uA2
    sigma2_ = 10*usigma2
    x20 = 20*(ux20-0.5)
    
    return A, sigma, x0, A2, sigma2_, x20
```

## Results
In this final seciton we demonstrate our results using `dynesty`'s `NestedSampler` for the single and double Gaussian cases.

### Single Gaussian

```python
dsampler = dynesty.NestedSampler(loglike1, prior_transform1, ndim=3,
                                    sample='rwalk')
dsampler.run_nested()
dres = dsampler.results

labels = [r'$A$', r'$sigma$', r'$x0$']

fig, axes = dyplot.traceplot(dsampler.results, labels=labels,
                             fig=plt.subplots(3, 2, figsize=(16, 12)))
fig.tight_layout()
```

```python
fig, axes = dyplot.cornerplot(dres,show_titles=True, 
                              title_kwargs={'y': 1.04}, labels=labels, color='dodgerblue',
                              fig=plt.subplots(3, 3, figsize=(15, 15)))
```

### Double Gaussian

```python
dsampler = dynesty.NestedSampler(loglike2, prior_transform2, ndim=6)
dsampler.run_nested()
dres = dsampler.results

labels = [r'$A$', r'$sigma$', r'$x0$', r'$A2$', r'$sigma2$', r'$x20$']

fig, axes = dyplot.traceplot(dsampler.results, labels=labels,
                             fig=plt.subplots(6, 2, figsize=(16, 12)))
fig.tight_layout()
```
